```python
#question 3
import requests
from bs4 import BeautifulSoup
```

```python
pge=requests.get('https://www.imdb.com/india/top-rated-indian-movies/?sort=ir,desc&mode=simple&page=1')
soup=BeautifulSoup(pge.content)
pge
```

```python
print(pge.text)
```

```python
rank=[]
title=[]
year=[]
for r in soup.find_all('td', class_='titleColumn'):
    rank.append(r.text.split('\n')[1])
    title.append(r.text.split('\n')[2])
    year.append(r.text.split('\n')[3])

    
print(rank)
print(title)
print(year)
```

```python
len(rank)
```

```python
len(title)
```

```python
len(year)
```

```python
imdb_rat=[]
for i in soup.find_all('td', class_='ratingColumn imdbRating'):
    imdb_rat.append(i.text.split('\n')[1])
    
imdb_rat
```

```python
import pandas as pd
imdb=pd.DataFrame({'Rank': rank, 'title': title, 'Year':year, 'Rating': imdb_rat})
imdb.head(100)
```

```python
#question 4
import requests
from bs4 import BeautifulSoup
```

```python
pg=requests.get('https://presidentofindia.nic.in/former-presidents.htm')
bs=BeautifulSoup(pg.content)
bs
```

```python
prez=[]
term=[]
for p in bs.find_all('div', class_='presidentListing'):
    prez.append(p.text.split('\n')[1])
    term.append(p.text.split('\n')[2])
print(prez)
print(term)
```

```python
import pandas as pd
president=pd.DataFrame({'President Name': prez, 'Term of office':term})
president
```
